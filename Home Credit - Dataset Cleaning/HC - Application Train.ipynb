```python
dfe = df.copy()
# Check duplicated value
duplicate_rows = dfe.duplicated()

# display number of duplicated row
print(f"number of duplicated row: {duplicate_rows.sum()}")

# Function to display columns contain nilai missing
def check_nan(dfe):
    null_values = dfe.isnull().sum()

    # create DataFrame for total and percentage of null 
    nvc = pd.DataFrame(null_values[null_values > 0].sort_values(), columns=['Total Null Values'])  
    nvc['Percentage'] = (nvc['Total Null Values'] / dfe.shape[0]) * 100
    nvc["Data Type"] = [dfe[col].dtype for col in nvc.index]  

    # order by total and percentage of null value 
    nvc.sort_values(by=["Total Null Values", "Percentage"], ascending=False, inplace=True)

    # show result
    return nvc.style.background_gradient(cmap='Blues')

styled_result = check_nan(dfe)

styled_result
```

```python
# 1. Drop columns with more than 50% missing values
missing_threshold = 0.5
cols_to_drop = dfe.columns[dfe.isnull().mean() > missing_threshold]
dfe.drop(columns=cols_to_drop, inplace=True)

# 2. Impute numeric columns with 10% - 50% missing values using the median
num_cols = dfe.select_dtypes(include=['float64', 'int64']).columns
for col in num_cols:
    missing_pct = dfe[col].isnull().mean()
    if 0.1 < missing_pct <= 0.5:
        dfe[col].fillna(dfe[col].median(), inplace=True)

# 3. Impute categorical columns with 10% - 50% missing values using the mode
cat_cols = dfe.select_dtypes(include='object').columns
for col in cat_cols:
    missing_pct = dfe[col].isnull().mean()
    if 0.1 < missing_pct <= 0.5:
        dfe[col].fillna(dfe[col].mode()[0], inplace=True)

# 4. Impute columns with less than 10% missing values
for col in dfe.columns:
    if dfe[col].isnull().mean() <= 0.1:
        if dfe[col].dtype == 'object':
            dfe[col].fillna(dfe[col].mode()[0], inplace=True)
        else:
            dfe[col].fillna(dfe[col].median(), inplace=True)

# 5. Check the final result
# Call the function and store the result
styled_result = check_nan(dfe)

# To display the result in a Jupyter Notebook, use:
styled_result
```

1. Drop columns with more than 50% missing values:
Columns with over half of their data missing are likely to lack sufficient information for meaningful analysis, and therefore are removed to maintain data quality.

2. Impute numeric columns with 1050% missing values using the median:
The median is robust to outliers, making it a safer choice for imputing missing values in skewed numerical distributions without distorting the data.

3. Impute categorical columns with 1050% missing values using the mode:
Using the mode preserves the dominant category in the data, maintaining the original distribution and reducing the risk of introducing bias.

4. Impute columns with less than 10% missing values using median/mode:
When the percentage of missing values is low, imputation using the median (for numerical) or mode (for categorical) can be done directly with minimal impact on the overall data structure.
